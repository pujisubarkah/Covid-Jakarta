```python
df_map_jakarta['kelurahan'] = df_map_jakarta['kelurahan'].str.upper()
```

```python
df_map_jakarta['kelurahan']
```

```python
merged = df_map_jakarta.set_index('kelurahan').join(df_covid_jakarta.set_index('nama_kelurahan'))
```

```python
merged
```

```python
variable = 'POSITIF'
vmin = merged['POSITIF'].min()
vmax = merged['POSITIF'].max()
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('off')
fig.set_facecolor('#ffffff')
fig.patch.set_facecolor('#ffffff')
ax.axis('off')
ax.margins(0)
ax.apply_aspect()
bbox = ax.get_window_extent().inverse_transformed(fig.transFigure)
w,h = fig.get_size_inches()
fig.set_size_inches(w*bbox.width, h*bbox.height)
merged.plot(column=variable, cmap='Reds', linewidth=0.3, ax=ax, edgecolor= '0.8')
cbar = fig.colorbar(plt.cm.ScalarMappable(cmap='Reds', norm=plt.Normalize(vmin=vmin, vmax=vmax)), orientation="vertical", aspect=20, shrink= 0.25, pad=0.1)
cbar.ax.tick_params(labelsize=6)
ax.set_title('Persebaran Covid-19 DKI Jakarta 11 September 2020',
              fontdict={'fontsize': '14',
                        'fontweight' : '5'})
ax.annotate('Sumber: https://corona.jakarta.go.id/',
           xy=(0.1, .1), xycoords='figure fraction',
           horizontalalignment='left', verticalalignment='bottom',
           fontsize=7, color='#555555')
plt.show()
```
